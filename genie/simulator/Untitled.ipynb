```python
import sys
import re
from itertools import chain
sys.path.append("../")
from pprint import pprint
from xml.etree.ElementTree import Element, tostring
from xml.dom.minidom import parseString
from textx.model import children_of_type, parent_of_type
```

```python
spike_exp = re.compile(
    "SPIKE : \t (?P<val>\d+\.*\d*)\t (?P<idvec>[0-9]+)\n")
start_exp = re.compile(
    "\[(?P<pid>\d+)\] NC = (?P<nc>\d+), SYN = (?P<syn>\d+), tmp_pre = (?P<tmp_pre>\d+), tmp_post = (?P<tmp_post>\d+)")
end_exp = re.compile(
    "\[(?P<pid>\d+)\] nsendmax=(?P<nsendmax>\d+) nsend=(?P<nsend>\d+) nrecv=(?P<nrecv>\d+) nrecv_useful=(?P<nrecv_useful>\d+)")
f = open("/Users/hashmup/genie/analyze/cluster/data/2018-02-02_18-04/tmp/job1.sh.o37548")
lines = f.readlines()
f.close()
_lines = []
start = {}
end = {}
spike = {}
maxid = 0
for i in range(len(lines)):
    line = lines[i]
#     print(m)
    m = start_exp.match(line)
    if m:
        pid = int(m.group("pid"))
        start[pid] = line
        maxid = max(maxid, pid)
        continue
    m = spike_exp.match(line)
    if m:
#         pid = int(m.group("pid"))
        idvec = int(m.group("idvec"))
        if idvec in spike:
            spike[idvec].append(line)
        else:
            spike[idvec] = [line]
        continue
    m = end_exp.match(line)
    if m:
        pid = int(m.group("pid"))
        end[pid] = line
if maxid > 0:
    maxid += 1
for pid in range(maxid):
    _lines.append(start[pid])
    _lines.append(end[pid])
for idvec in spike:
    for line in spike[idvec]:
        _lines.append(line)
print(_lines)
```
